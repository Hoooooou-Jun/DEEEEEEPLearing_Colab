```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
# urllib.request 라이브러리를 이용해 github 저장소의 데이터를 불러온다.
import urllib.request
DOWNLOAD = "https://raw.githubusercontent.com/yoonmh/AI-Data/main/AI(1)/"
filename = "Advertising.csv"
url = DOWNLOAD + filename
urllib.request.urlretrieve(url, filename)
```

```python
# 광고 데이터를 Pandas의 read_csv() 메소드로 읽는다.
# head() 메소드로 처음 5줄을 표로 나타낸다.
data = pd.read_csv('Advertising.csv')
print(data.shape)
data.head()

# TV: TV 광고에 사용한 비용
# Radio : 라디오 광고에 사용한 비용
# Newspaper : 신문 광고에 사용한 비용
# Sales : 매출($1,000)
```

```python
# 데이터 전처리
# 의미없는 첫번째 (Unnamed: 0) 열은 삭제한다.

data.drop('Unnamed: 0',axis = 1, inplace = True)
```

```python
# TV, 라디오, 신문 광고에 대한 매출효과를 그래프로 표시
import seaborn as sns

sns.pairplot(data, x_vars=['TV','Radio','Newspaper'],
             y_vars='Sales', 
             height=3.5,
             kind='reg')
plt.show()
```
